### Task 1:
Consider the following experiment:
1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_{i}, i = 1, 2,...,k$ and remains the same from trial to trial. Notice that $p_{1} + p_{2} + p_{3} + ... + p_{k} = 1$.
4. The trials are independent.
5. The random variables of interest are $Y_{1}, Y_{2},...,Y_{k}$, where $Y_{i}$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_{1} + Y_{2} + Y_{3} + ... + Y_{k} = n$.
The joint probability distribution $P(y) = \frac{n!}{y_{1}!y_{2}!y_{3}!...y_{n}!} P_{1}^{y_{1}} P_{2}^{y_{2}} P_{3}^{y_{3}} ... P_{n}^{y_{n}}$ 

#### Question 1:
Show that the expanded value $E[y] = \sum y \cdot P(y) = n \cdot p_{i}$

### Answer:
Substitute:<br><br>
Given $P(y) = \frac{n!}{y_{1}!y_{2}!y_{3}!...y_{k}!} P_{1}^{y_{1}} P_{2}^{y_{2}} P_{3}^{y_{3}} ... P_{k}^{y_{k}}$, where $Y_{1} + Y_{2} + Y_{3} + ... + Y_{k} = n$ and $p_{1} + p_{2} + p_{3} + ... + p_{k} = 1$, substitute this into the formula:

\begin{align*}
E[Y_{i}] = \Sigma_{y} y \cdot \frac {n!}{y_{1}!y_{2}!y_{3}!...y_{k}!} \cdot P_{1}^{y_{1}} P_{2}^{y_{2}} P_{3}^{y_{3}} ... P_{k}^{y_{k}}
\end{align*}

<br>Simplify the Sum:<br>
Since $Y_{i}$ is the number of trials for which the outcome falls into cell $i$, in each term of the sum only one $y_{i}$ term will be non-zero and the rest will be zero. Therefore, the sum simplified to:
\begin{align*}
E[Y_{i}] = \Sigma_{y_{i}} y_{i} \cdot \frac {n!}{y_{i}!(n - y_{i})!} p_{i}^{y_{i}} (1-p_{i})^{n-y_{i}}
\end{align*}

<br>
The simplified sum represents the expected value of a binomial random variable with parameters $n$ and $p_{i}$. Thus, the expected value is proven:
\begin{align*}
E[Y_{i}] = n \cdot p_{i}
\end{align*}

If 500 adults are sampled randomly, find the probability rhat the sample contains 100 person between 18 and 24, 200 between, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64.. What is the expected value for to obtain a person in the 65 and above?

The probability for each group:
<ul>18-24 group: 0.18</ul>
<ul>25-34 group: 0.23</ul>

The probability that the sample contains the given number of people among the groups can be calculated using the formula:
\begin{align*}

### Task 2: 
Consider the following experiment:
1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to $q = (1 - p)$.
4. The trials are independent.
5. The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

#### Question 1:
From the following steps shown above, derive the probability distribution of the experiment.
### Answer:
The probability distribution of $Y$ can be derived using the binomial distribution formula:
\begin{align*}
P(Y = y) = (^{n}_{y}) \cdot p^{y} \cdot q^{n-y}
\end{align*}
where:
<ul>$(^{n}_{y})$ is the binomial coefficient, which represents the number of ways to choose $y$ successes from $n$ trials.</ul>
<ul>$p^{y}$ is the probability of $y$ successes occuring</ul>
<ul>$q^{n-y}$ is the probability of $n-y$ failures occuring</ul>

#### Question 2:
Show that the expectation of this probability distribution is $E[y] = \Sigma y\cdot P(y) = np$.
### Answer:
Calculate $P(y)$ using the binomial distribution formula,
\begin{align*}
P(Y=y) = (^{n}_{y}) \cdot p^{y} \cdot q^{n-y}
\end{align*}
Multiply each $y$ by its corresponding $P(y)$.<br>
$y \cdot P(y)$ <br>
Sum up all these products for each possible $y$ value from $0$ to $n$:
\begin{align*}
\Sigma ^{n} _{y=0} y\cdot P(y)
\end{align*}
Since $P(y)$ follows the binomial distribution, we have:
\begin{align*}
E[Y] = \Sigma^{n}_{y=0} y \cdot P(y) = \Sigma^{n}_{y=0} y \cdot (^{n}_{y}) \cdot p^{y} \cdot q^{n-y}
\end{align*}

After simplifying we arrive at:
\begin{align*}
E[Y] = np
\end{align*}

### Task 3:
Consider a probility distribution of $p(y) = \frac{\lambda^{y}}{y!}e^{-\lambda}$.

#### Question 1: 
Find a general formula for the expected value of this distribution.

### Answer:
The expected value $E[Y]$ of this distribution can be found using the formula:
\begin{align*}
E[Y] = \Sigma ^{\infty}_{y=0} y\cdot P(y)
\end{align*}

Substitute the given probability distribution $P(y) = \frac{\lambda^{y}}{y!}e^{-\lambda}$ into the formula:
\begin{align*}
E[Y] = \Sigma ^{\infty}_{y=0} y\cdot \frac{\lambda^{y}}{y!}e^{-\lambda}
\end{align*}

Simplify the expression:
\begin{align*}
E[Y] = \Sigma ^{\infty}_{y=1}\cdot \frac{\lambda^{y}}{y-1!}e^{-\lambda}
\end{align*}

Substitute $k = y - 1$:
\begin{align*}
E[Y] = \Sigma ^{\infty}_{k=0}\cdot \frac{\lambda^{k+1}}{k!}e^{-\lambda}
\end{align*}

Factor out $\lambda$ from the sum:
\begin{align*}
E[Y] = \lambda \Sigma ^{\infty}_{k=0}\cdot \frac{\lambda^{k}}{k!}e^{-\lambda}
\end{align*}


The sum $\Sigma ^{\infty}_{k=0}\cdot \frac{\lambda^{k}}{k!}$ is the Taylor series expansion of $e^{\lambda}$:

\begin{align*}
E[Y] = \lambda \cdot e^{\lambda} \cdot e^{-\lambda}
\end{align*}


Simplifying gives us:
\begin{align*}
E[Y] = \lambda
\end{align*}

Thus, the general formula for the expected value of this distribution is $E[Y] = \lambda$

#### Question 2:
A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.


### Answer:
Given $\lambda = 5$ seedlings per square yard and the forester locates ten 1-square-yard sampling regions, the distribution follows a Poisson distribution with parameter $\lambda = 5$ as the mean is equal to the parameter for a Poisson distribution.<br><br>
\begin{align*}
P(Y=y) = \frac{e^{-\lambda}\lambda^{y}}{y!}
\end{align*}

To find the probability that none of the regions will contain seedlings, substitute $Y = 0$ and $\lambda = 5$ into the Poisson distribution formula:
\begin{align*}
P(Y=0) = \frac{e^{-5} \cdot 5^{y}}{0!}
\end{align*}

\begin{align*}
P(Y=0) = e^{-5} \approx 0.00674
\end{align*}

The probability that none of the regions will contain seedlings is approximately $0.00674$ or $0.674\%$
