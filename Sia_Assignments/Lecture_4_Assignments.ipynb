#### 1. Kruskal's Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
plt.figure(figsize=(10, 8))

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edges = [("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15), ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35), ("F", "G", 50), ("G", "E", 45)]

G.add_weighted_edges_from(edges)
mst = nx.minimum_spanning_tree(G, algorithm='kruskal')
pos = nx.planar_layout(G)

nx.draw(G, pos, with_labels=True, node_color='pink')
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
nx.draw_networkx_edges(G, pos, edge_color='black', width=1)
nx.draw_networkx_edges(mst, pos, edge_color='fuchsia', width=2)

plt.show()
```

#### 2. Prim's Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
plt.figure(figsize=(10, 8))

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edges = [("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15), ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35), ("F", "G", 50), ("G", "E", 45)]

G.add_weighted_edges_from(edges)
mst = nx.minimum_spanning_tree(G, algorithm='prim')
pos = nx.planar_layout(G)

nx.draw(G, pos, with_labels=True, node_color='pink')
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
nx.draw_networkx_edges(G, pos, edge_color='black', width=1)
nx.draw_networkx_edges(mst, pos, edge_color='fuchsia', width=2)

plt.show()
```

#### 3. Djitkraka's Shortest Path Algorithm

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
plt.figure(figsize=(10, 8))

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edges = [("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15), ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35), ("F", "G", 50), ("G", "E", 45)]

G.add_weighted_edges_from(edges)

start = "A"
end = "G"

shortest_path = nx.dijkstra_path(G, source=start, target=end)
shortest_path_edges = list(zip(shortest_path, shortest_path[1:]))

pos = nx.planar_layout(G)

nx.draw(G, pos, with_labels=True, node_color='pink')
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
nx.draw_networkx_edges(G, pos, edge_color='black', width=1)
nx.draw_networkx_edges(G, pos, edgelist=shortest_path_edges, edge_color='fuchsia', width=2)

plt.show()
```
