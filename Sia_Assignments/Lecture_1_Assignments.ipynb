### Problem 1:
Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e$ = $\sum^{\infty}_{k=0} \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k-1)(k-2)(k-3)...,\forall k \in \mathbb{Z}^{+}$, note that $0! = 1$. Furthermore, a sum notation $\sum^{\infty}_{k=0} k = 0 + 1 + 2 + 3 + ... + ...$

### Answer:
Proof by contradiction<br>
#### Step 1:
Suppose $e$ is rational. This means it can be expressed as a fraction in its simplest form,
\begin{align*}
e = \frac{p}{q}
\end{align*}

Where $p$ and $q$ are integers with no common factors, and $q \neq 0$.

#### Step 2:
Consider the partial sum $S_{n}$ of the series up to $n$ terms:
\begin{align*}
S_{n} = \sum^{n}_{k=0} \frac{1}{k!}
\end{align*}
Define the remainder $R_{n}$ of the series as:
\begin{align*}
R_{n} = e - S_{n} = \sum^{\infty}_{k=n-1} \frac{1}{k!}
\end{align*}

#### Step 3:
The remainder $R_{n}$ can be bounded by: 
\begin{align*}
R_n < \frac{(1}{(n+1)!} + \frac{1}{(n+2)!} + \cdots < \frac{1}{(n+1)!} \left(1 + \frac{1}{n+1} + \frac{1}{(n+1)^{2}} + \cdots \right)
\end{align*}

The series in parentheses is a geometric series with sum less than 2, hence:
\begin{align*}
R_n < \frac{2}{(n+1)!}
\end{align*}

#### Step 4:
Choose $n$ such that $(n + 1)! > 2q$.
\begin{align*}
R_n < \frac{2}{(n+1)!} < \frac{2}{2q} = \frac{1}{q}
\end{align*}

#### Step 5:
Since $S_n$ is a sum of rational numbers, $S_n$ is rational.
\begin{align*}
e=S_{n} + R_{n} = \frac{a}{b} + R_{n}
\end{align*}

for some integers $a$ and $b$. Multiplying both sides by $q$ gives:
\begin{align*}
qe = qS_{n} + qR_{n} 
\end{align*}

Since $qe = p$ and $qS_{n}$ is an integer, let $qS_{n} = m$ where $m$ is an integer
\begin{align*}
p = m + qR_{n} 
\end{align*}

Thus, $qR_{n} = p - m$ must be an integer.

#### Step 6:
Since $R_{n} < \frac{1}{q}$, we have:
\begin{align*}
0 < qR_{n} < 1
\end{align*}
which means $qR_{n}$ is a positive number less than 1, contradicting the fact that $qR_{n}$ must be an integer.<br>

Since our assumption that $e$ is rational leads to a contradiction, this means Euler's number $e$ is irrational:
\begin{align*}
e \notin \mathbb{Q}
\end{align*}

### Problem 2:
Prove Minkowski's Inequality for sums, $\forall p > 1, (a_{k}, b_{k}) > 0$:

\begin{align*} 
        [ \Sigma^{n}_{k=1}|a_{k} + b_{k}|^{p}]^\frac{1}{p} \leq [ \Sigma^{n}_{k=1}|a_{k}|^{p}]^\frac{1}{p} + [ \Sigma^{n}_{k=1}|b_{k}|^{p}]^\frac{1}{p}
\end{align*}

### Answer:<br>
#### Step 1:
For $p > 1$, define $q$ by:
\begin{align*} 
q = \frac{p}{p - 1}
\end{align*}

Introduce the Hölder's Inequality
\begin{align*} 
\Sigma^{n}_{k=1} |x_{k}y_{k}| \leq (\Sigma^{n}_{k=1}|x_{k}|^{p})^{\frac{1}{p}}(\Sigma^{n}_{k=1}|y_{k}|^{q})^\frac{1}{q}
\end{align*}

#### Step 2:
Apply Hölder's Inequality
\begin{align*} 
    x_{k} = |a_{k} + b_{k}|^{p-1}
\end{align*}
\begin{align*}
    y_{k} = |a_{k} + b_{k}|
\end{align*}

#### Step 3:
Use the triangle inequality on the term $|a_{k} + b_{k}|$ within $y_{k}$:
\begin{align*}
|a_{k} + b_{k}| = |(a_{k} + b_{k}) - 0| \leq |a_{k}| + |b_{k}|
\end{align*}

Therefore:
\begin{align*}
y_{k} = |(a_{k} + b_{k}) \leq |a_{k}| + |b_{k}|
\end{align*}

#### Step 4:
Substitute into Hölder's Inequality
\begin{align*}
\Sigma^{n}_{k=1} |a_{k} + b_{k}|^{p} \leq [\Sigma^{n}_{k=1} |a_{k} + b_{k}|^{p-1} |a_{k} + b_{k}|]^\frac{1}{p} \cdot [\Sigma^{n}_{k=1} (|a_{k}| + |b_{k}|)^{q}]^\frac{1}{q}
\end{align*}

#### Step 5:
Simplify the terms:
\begin{align*}
[\Sigma^{n}_{k=1} (|a_{k}| + |b_{k}|)^{q}]^\frac{1}{q} = [\Sigma^{n}_{k=1} |a_{k}|^{q}]^\frac{1}{q} + [\Sigma^{n}_{k=1}  |b_{k}|^{q}]^\frac{1}{q}
\end{align*}

Put it all together:
\begin{align*}
[\Sigma^{n}_{k=1} |a_{k} + b_{k}|^{p}] \leq [\Sigma^{n}_{k=1} |a_{k} + b_{k}|^{p}]^\frac{1}{p} \cdot [\Sigma^{n}_{k=1} |a_{k}|^{q}]^\frac{1}{q} + [\Sigma^{n}_{k=1}  |b_{k}|^{q}]^\frac{1}{q}
\end{align*}

Divide both sides by the common term on the left:
\begin{align*}
1 \leq [\Sigma^{n}_{k=1} |a_{k}|^{q}]^\frac{1}{q} + [\Sigma^{n}_{k=1} |b_{k}|^{q}]^\frac{1}{q}
\end{align*}

Take the $pth$ power of both sides:
\begin{align*}
1^{p} \leq [\Sigma^{n}_{k=1} |a_{k}|^{q}] + [\Sigma^{n}_{k=1} |b_{k}|^{q}]
\end{align*}

The key terms can be rewritten using the definition of q:
\begin{align*}
[\Sigma^{n}_{k=1} |a_{k}|^{q}] = [\Sigma^{n}_{k=1} |a_{k}|^\frac{p}{p-1}] 
\end{align*}

And after doing the same for the term with $b_{k}$, we raise both sides to the power of $\frac{1}{p}$ to get:
\begin{align*}
[ \Sigma^{n}_{k=1}|a_{k} + b_{k}|^{p}]^\frac{1}{p} \leq [ \Sigma^{n}_{k=1}|a_{k}|^{p}]^\frac{1}{p} + [ \Sigma^{n}_{k=1}|b_{k}|^{p}]^\frac{1}{p}
\end{align*}

### Problem 3:
Prove the triangle inequality $|x + y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$
### Answer:
#### Case 1: $(x \geq 0, y \geq 0)$
Both $x$ and $y$ are positive or zero. Since the sum of non-negative numbers is non-negative:
\begin{align*}
(x + y) \geq 0
\end{align*}
Since the absolute value of non-negative numbers remains the same, we can write:
\begin{align*}
|x + y| = x + y or |x| + |y| = x + y
\end{align*}
Thus, $|x + y| = |x| + |y|$

#### Case 2: $(x \geq 0, y < 0)$
$x$ is positive or zero and $y$ is negative. Since the absolute value of a negative number is positive:
\begin{align*}
|x + y| \leq x - y 
\end{align*}
Consider that $x$ and $y$ have opposite signs, $|x + (-y)| \leq |x| + |y|$

#### Case 3: $(x < 0, y \geq 0)$
Just like case 2, x and y have opposite signs.
\begin{align*}
|x + y| \leq -x + y 
\end{align*}
Thus, $|-x + y| \leq |x| + |y|$

#### Case 4: $(x < 0, y < 0)$
Since both are non-negative, we can write:
\begin{align*}
|x + y| = -(x + y) 
\end{align*}
\begin{align*}
|x| + |y| = -x -y 
\end{align*}
Therefore, $|x + y| = |x| + |y|$
#### Conclusion:
In all four cases, we arrived at an inequality that holds true

### Problem 4:
Prove Sedrakayan's Lemma $\forall u_{i}, v_{i} \in \mathbb{R}^{+}$:
\begin{align*}
\frac{(\Sigma^{n}_{i=1} u_{i})^{2}}{\Sigma^{n}_{i=1} v_{i}} \leq \Sigma^{n}_{i=1} \frac{(u_{i})^{2}}{v_{i}}
\end{align*}
### Answer:
#### Step 1:
The Cauchy-Schwarz inequality states that for any sequences of real numbers $a_{i}$ and $b_{i}$,
\begin{align*}
(\Sigma ^{n}_{i = 1} a_{i}b_{i})^{2} \leq (\Sigma ^{n}_{i = 1} a_{i}^{2})(\Sigma ^{n}_{i = 1} b_{i}^{2})
\end{align*}
Apply Cauchy-Schwarz inequality:
\begin{align*}
a_{i} = \frac{u_{i}}{\sqrt{v_{i}}}   \enspace and \enspace   b_{i} = \sqrt{v_{i}}
\end{align*}
Now we get,
\begin{align*}
(\Sigma ^{n}_{i = 1} \frac{u_{i}}{\sqrt{v_{i}}}\cdot \sqrt{v_{i}})^{2} \leq (\Sigma ^{n}_{i = 1} (\frac{u_{i}}{\sqrt{v_{i}}})^{2})(\Sigma ^{n}_{i = 1} (\sqrt{v_{i}})^{2})
\end{align*}

#### Step 2:
Simplify the expressions:
\begin{align*}
(\Sigma ^{n}_{i = 1} \frac{u_{i} \sqrt{v_{i}}}{\sqrt{v_{i}}})^{2} = (\Sigma ^{n}_{i = 1}u_{i})^{2}
\end{align*}
\begin{align*}
(\Sigma ^{n}_{i = 1} (\frac{u_{i}}{\sqrt{v_{i}}})^{2})(\Sigma ^{n}_{i = 1} (\sqrt{v_{i}})^{2}) = (\Sigma ^{n}_{i = 1} \frac{u_{i}^{2}}{v_{i}})(\Sigma ^{n}_{i = 1} v_{i})
\end{align*}

#### Step 3:
Combine the results:
\begin{align*}
(\Sigma ^{n}_{i = 1}u_{i})^{2} \leq (\Sigma ^{n}_{i = 1} \frac{u_{i}^{2}}{v_{i}})(\Sigma ^{n}_{i = 1} v_{i})
\end{align*}

#### Step 4: 
Divide both sides by $\Sigma ^{n}_{i = 1} v_{i}$ to get:
\begin{align*}
\frac{(\Sigma ^{n}_{i = 1}u_{i})^{2}}{\Sigma ^{n}_{i = 1} v_{i}} \leq \Sigma ^{n}_{i = 1} \frac{(u_{i})^{2}}{v_{i}}
\end{align*}
This completes the proof.

1. Consider the function $f(x) = 2x + 1, g(x) = 2y - 1, j(x) = |f(x) + g(x)|$, and $h(x) = j(x)^{2}$, show for the following inequality that:

\begin{align*}
\sqrt{|f(x)|^{2}} \geq \sqrt{h(x)} - \sqrt{|g(x)|^{2}}, \forall (x, y) \in \mathbb{Z}^{+}
\end{align*}

### Answer:
#### Step 1:
Break it down:
\begin{align*}
f(x) = 2x + 1
\end{align*}
\begin{align*}
g(x) = 2y - 1
\end{align*}
\begin{align*}
j(x) = |f(x) + g(x)| = |(2x + 1)+(2y - 1)| = |2x + 1 + 2y - 1| = |2(x+y)|
\end{align*}
\begin{align*}
h(x) = j(x)^{2} = (2|x+y|)^{2} = 4(x+y)^2
\end{align*}

#### Step 2:
Substitute and evaluate:
\begin{align*}
\sqrt{|f(x)|^{2}} = \sqrt{(2x + 1)^{2}} = |2x + 1|
\end{align*}
\begin{align*}
\sqrt{h(x)} = \sqrt{4(x + y)^{2}} = 2|x + y|
\end{align*}
\begin{align*}
\sqrt{|g(x)|^{2}} = \sqrt{(2y - 1)^{2}} = |2y - 1|
\end{align*}

Combine the terms:
\begin{align*}
|2x + 1| \leq 2|x + y| - |2y - 1|
\end{align*}
#### Step 3:
#### Case: $x + y \geq 1$:
If $2x + 1 \geq 0$ and $2y - 1 \geq 0$, 
\begin{align*}
|2x + 1| = 2x + 1 \enspace and \enspace  |2y - 1| = 2y - 1
\end{align*}
\begin{align*}
2x + 1 = \geq 2(x + y) - (2y - 1)
\end{align*}
\begin{align*}
2x + 1 = \geq 2x + 1
\end{align*}
This holds true as equality.
#### Conclusion:
The inequality holds true by considering the absolute values.
