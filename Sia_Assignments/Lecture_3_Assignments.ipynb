### Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
seed =  88
n = 4  
m = 2  

# Generate a random graph with n nodes and m edges
G = nx.gnm_random_graph(n, m, seed=seed)

G.add_nodes_from([0, 1, 2, 3])

edgelist = [(0,1),(0,2),(1,3), (1,1),(2,0),(3,0)]

G.add_edges_from(edgelist)

pos = nx.spectral_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
plt.show()
```

A graph consists of a set of vertices (nodes) and a set of edges. An edge is the line connecting two vertices. The above figure is an undirected graph. The edges in this type of graph simply link the vertices.

### Directed Graphs (Digrahs)

```python
import matplotlib.pyplot as plt
import networkx as nx

# Generate a directed graph 
G = nx.DiGraph()

G.add_nodes_from([1, 2, 3])

edgelist = [(1,2),(2,3),(3,1)]

G.add_edges_from(edgelist)

pos = nx.circular_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
plt.show()
```

Similar to undirected graphs, directed graphs consist of vertices and edges. The key difference with directed graphs is that their edges have arrows to indicate the direction of the relationship between vertices.

### Complete Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.gnm_random_graph(n, m)

G.add_nodes_from([0, 1, 2, 3])

edgelist = [(0,1),(0,2),(1,2),(1,3),(2,3),(3,0)]

G.add_edges_from(edgelist)

pos = nx.shell_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
plt.show()
```

A complete graph where all its vertices are connected to every other vertex and there are no two edges connected the same two pairs of vertices.

### Bipartite Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from(['Cat','Rabbit','Hawk'], bipartite=0)
G.add_nodes_from(['Berries','Carrots'], bipartite=1)
G.add_edges_from([('Cat','Berries'),('Cat','Carrots'),('Rabbit','Berries'), ('Rabbit','Carrots'),('Hawk','Berries'),('Hawk','Carrots')])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, node_color=['pink','lightblue','peru','mediumpurple','orange'])
plt.show()
```

Bipartite graph or bigraph is a graph which shows two sets of objects. Each set contains objects which do not connect with one another but each object connects with every other object from the other set.

### Subgraphs

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.gnm_random_graph(n, m)
SG = nx.gnm_random_graph(n, m)

G.add_nodes_from([0, 1, 2, 3])
SG.add_nodes_from(['a', 'b', 'c'])

edgelist = [(0,1),(0,2),(1,2),(1,3),(2,3),(3,0)]
subedgelist = [('a','b'),('b','c'),('c','a')]

G.add_edges_from(edgelist)
SG.add_edges_from(subedgelist)

pos = nx.shell_layout(G)
pos = nx.shell_layout(SG)

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
nx.draw(SG, pos=pos, with_labels=True, node_color ='pink')

plt.show()
```

Subgraphs are small sections of a bigger graph.

### Degree of a Vertex

### Travelling in Graphs

#### Walk

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 88

G = nx.Graph()

G.add_nodes_from(['a','b','c','d','e'])
edgelist = [('a', 'b'), ('a', 'd'), ('b', 'c'), ('b', 'd'), ('b', 'e'), ('c', 'd'), ('d', 'e')]

G.add_edges_from(edgelist)

pos = nx.shell_layout(G)

x = [('a','b'), ('b','d'), ('d','a'), ('a','b'), ('b','c')]

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
nx.draw_networkx_edges(G, pos=pos, edgelist=x, edge_color='darkgreen', width=2)

plt.show()
```

Walk is a sequence of vertices and edges that starts at one vertex and ends at another or the same vertex. Walks allow us to visit the same vertex and edge more than once.

### Connected Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 88

G = nx.Graph()
G2 = nx.Graph()

G.add_nodes_from(['a','b','c','d'])
edgelist = [('a', 'b'), ('a', 'c'), ('b', 'c'), ('d', 'a')]

G2.add_nodes_from(['A','B','C'])
edgelist2 = [('A', 'B')]

G.add_edges_from(edgelist)
G2.add_edges_from(edgelist2)

pos = nx.shell_layout(G)
pos2 = nx.random_layout(G2)

plt.figure(figsize=(3,3))
nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
plt.title("Connected")

plt.figure(figsize=(3,3))
nx.draw(G2, pos=pos2, with_labels=True, node_color = 'aquamarine')
plt.title("Not connected")
plt.show()
```

A connected graph is a graph where every pair of vertices are connected by an edge and there exists at least 1 path from one vertex to another.

### Hamiltonian Circuits

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.gnm_random_graph(n, m)

G.add_nodes_from([0, 1, 2, 3, 4])

edgelist = [(0,1),(1,2),(2,3),(3,4),(4,0)]

G.add_edges_from(edgelist)

pos = nx.spring_layout(G)

nx.draw(G, pos=pos, with_labels=True, node_color = 'pink')
plt.show()
```

Hamiltonian circuit is a circuit that must pass through every vertex once. It starts and ends at the same vertex.
